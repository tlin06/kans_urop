```python
import matplotlib.pyplot as plt
import pickle
```

```python
with open('data/grad_comparisons_data.p', 'rb') as fp:
    data = pickle.load(fp)
```

```python
epochs = [n for n in range(300)]
methods = ['auto pytorch', 'manual list', 'manual values']
for m in methods:
    plt.plot(epochs, data[(20, 1, 'SGD', m)], label=m)
plt.legend(loc = 'best')
plt.show()
```
