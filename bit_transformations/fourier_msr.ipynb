```python
def dft_mag(x):
    x = (-1) ** np.array(x)
    return np.abs(np.fft.rfft(x)[1:])

def psd(x):
    return dft_mag(x) ** 2
def norm_psd(x):
    vec = psd(x)
    return vec / np.linalg.norm(vec)
# note all the data points have dc component = 0
```
